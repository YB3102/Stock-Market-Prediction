### Data Collection 

Collecting stock prices for companies in the S&P 500 list. 
Using yfinance api / library


```python
import yfinance as yf
import pandas as pd
import requests
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_S%26P_500_companies'

tables = pd.read_html(url)

sp500_table = tables[0]

tickers = sp500_table['Symbol'].tolist()

for i in range(len(tickers)):
    if tickers[i] == "BF.B":
        tickers[i] = "BF-B"
    elif tickers[i] == "BRK.B":
        tickers[i] = "BRK-B"
```

```python
# Define the start and end dates for the historical data
start_date = '2020-01-01'
end_date = '2024-11-12'

# Download the data
data = yf.download(tickers, start=start_date, end=end_date, group_by='ticker')
```

```python
data.head()
# Filter to keep only the 'Adj Close' column for each ticker
adj_close_data = data.xs('Adj Close', level=1, axis=1)

# Display the first few rows of the adjusted close prices
adj_close_data.head()
```

```python
from pandas import json_normalize

articles = data.get("feed", [])

# Flattening the JSON data in 'feed' and creating a DataFrame
articles_df = json_normalize(
    articles,
    record_path='topics',  # Explode the 'topics' field into rows
    meta=[
        'title', 'url', 'time_published', 'summary', 'source', 'source_domain', 
        'overall_sentiment_score', 'overall_sentiment_label'
    ],
    errors='ignore'  # Ignore errors if certain fields are missing
)

# Displaying the DataFrame
print(articles_df.head())
```
